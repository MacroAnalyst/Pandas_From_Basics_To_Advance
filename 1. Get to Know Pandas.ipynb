<h3>People</h3><br>
<strong>ID</strong>: Customer's unique identifier<br>
<strong>Year_Birth</strong>: Customer's birth year<br>
<strong>Educatio</strong>n: Customer's education level<br>
<strong>Marital_Status</strong>: Customer's marital status<br>
<strong>Income</strong>: Customer's yearly household income<br>
<strong>Kidhom</strong>e: Number of children in customer's household<br>
<strong>Teenhome</strong>: Number of teenagers in customer's household<br>
<strong>Dt_Customer</strong>: Date of customer's enrollment with the company<br>
<strong>Recency</strong>: Number of days since customer's last purchase<br>
<strong>Complain</strong>: 1 if customer complained in the last 2 years, 0 otherwise<br>

<h3>Products</h3><br>
<strong>MntWines</strong>: Amount spent on wine in last 2 years<br>
<strong>MntFruits</strong>: Amount spent on fruits in last 2 years<br>
<strong>MntMeatProducts</strong>: Amount spent on meat in last 2 years<br>
<strong>MntFishProducts</strong>: Amount spent on fish in last 2 years<br>
<strong>MntSweetProducts</strong>: Amount spent on sweets in last 2 years<br>
<strong>MntGoldProds</strong>: Amount spent on gold in last 2 years<br>

<h3>Promotion</h3><br>
<strong>NumDealsPurchases</strong>: Number of purchases made with a discount<br>
<strong>AcceptedCmp1</strong>: 1 if customer accepted the offer in the 1st campaign, 0 otherwise<br>
<strong>AcceptedCmp2</strong>: 1 if customer accepted the offer in the 2nd campaign, 0 otherwise<br>
<strong>AcceptedCmp3</strong>: 1 if customer accepted the offer in the 3rd campaign, 0 otherwise<br>
<strong>AcceptedCmp4</strong>: 1 if customer accepted the offer in the 4th campaign, 0 otherwise<br>
<strong>AcceptedCmp5</strong>: 1 if customer accepted the offer in the 5th campaign, 0 otherwise<br>
<strong>Response</strong>: 1 if customer accepted the offer in the last campaign, 0 otherwise<br>

<h3>Place</h3><br>
<strong>NumWebPurchases</strong>: Number of purchases made through the company’s web site<br>
<strong>NumCatalogPurchases</strong>: Number of purchases made using a catalogue<br>
<strong>NumStorePurchases</strong>: Number of purchases made directly in stores<br>
<strong>NumWebVisitsMonth</strong>: Number of visits to company’s web site in the last month<br>

## <font face="gotham" color="purple"> A More Complicated Example </font>

Find the consumers who are 
1. single with at master's degree or above
2. with at least a kid or teen at home
3. income between $45000$ and $65000$
4. spent on wine more than $50$
5. last purchase within $30$ days

```python
is_PhD_or_Master = campaign['Education'].isin(['Phd', 'Master'])
campaign_sg_mst_phd = campaign[(campaign['Marital_Status'] == 'Single') & is_PhD_or_Master]
campaign_sg_mst_phd.head()
```

```python
campaign_sg_mst_phd_kid_tean = campaign_sg_mst_phd[(campaign_sg_mst_phd['Kidhome'] > 0) & (campaign_sg_mst_phd['Teenhome'] > 0)]
campaign_sg_mst_phd_kid_tean.head()
```

```python
campaign_sg_mst_phd_kid_tean_income = campaign_sg_mst_phd_kid_tean[(campaign_sg_mst_phd_kid_tean['Income']>45000) &(campaign_sg_mst_phd_kid_tean['Income']<65000)]
campaign_sg_mst_phd_kid_tean_income.head()
```

```python
campaign_sg_mst_phd_kid_tean_income_amtwine = campaign_sg_mst_phd_kid_tean_income[campaign_sg_mst_phd_kid_tean_income['MntWines']>50]
campaign_sg_mst_phd_kid_tean_income_amtwine
```

```python
campaign_sg_mst_phd_kid_tean_income_amtwine[campaign_sg_mst_phd_kid_tean_income_amtwine['Recency']<30]
```
